```python
from google.colab import drive
drive.mount('/content/gdrive')
df_train = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/dz5ML/Train.csv')
df_test = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/dz5ML/Test.csv')
df_train.head(20)
```

```python
# Удалим ненужные колонки
df_train = df_train.drop(columns=['id', 'date', 'n_photos'])
y_df_test_id = df_test.id.values
df_test = df_test.drop(columns=['id', 'date', 'n_photos'])
```

```python
# Смотрим, где есть NaN
df_train.isna().any()
```

```python
# Убираем NaN в build_tech и в g_lift
df_train.build_tech = df_train.build_tech.fillna(value=df_train.build_tech.median())
df_train.metro_dist = df_train.metro_dist.fillna(value=df_train.metro_dist.mean())
df_train.g_lift = df_train.g_lift.fillna(value=df_train.g_lift.median())

df_test.build_tech = df_test.build_tech.fillna(value=df_test.build_tech.median())
df_test.metro_dist = df_test.metro_dist.fillna(value=df_test.metro_dist.mean())
df_test.g_lift = df_test.g_lift.fillna(value=df_test.g_lift.median())
```

```python
# Приведем build_tech и g_lift к типу int64
df_train = df_train.astype({'build_tech' : 'Int64', 'g_lift' : 'Int64'})
df_test = df_test.astype({'build_tech' : 'Int64', 'g_lift' : 'Int64'})
```

```python
df_train = df_train[df_train.price != 1000000]
```

```python
plt.figure(figsize=(20, 8))
plt.hist(df_train.price, bins=500);
```

Выброс убран!

```python
y = df_train.price.values
X_train = df_train.drop(columns='price').values
```

```python
scaler = StandardScaler()
X_train_scaler = scaler.fit_transform(X_train)
```

```python
quadratic = PolynomialFeatures(degree=2)
X_quad = quadratic.fit_transform(X_train_scaler)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_quad, y, test_size=0.2, random_state=42)
```

```python
XGBR = xgb.XGBRegressor(objective='reg:squarederror', colsample_bytree=0.5, colsample_bynode=0.5, n_jobs=15, colsample_bylevel=0.5, booster='gbtree', seed=42,
                        learning_rate=0.65, min_child_weight=0.001, n_estimators=1000)

LGBM = LGBMRegressor(random_state=42, reg_alpha=0.05, reg_lambda=0.05, n_jobs=16, n_estimators=1000,
                      learning_rate=0.1, num_leaves=1000)

catboost = CatBoostRegressor(loss_function="RMSE", score_function='L2', n_estimators=1000, learning_rate=0.3)
```

```python
voting = VotingRegressor([('LGBM', LGBM), ('XGBR', XGBR), ('catboost', catboost)])
```

```python
voting.fit(X_train, y_train)
```

```python
y_pred = voting.predict(X_test)
```

```python
mean_absolute_error(y_test, y_pred)
```
